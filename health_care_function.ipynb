```python
import pandas as pd
import perprocess

class HealthCareExcel():
    def __init__(self,excel_location): 
        self.excel_location = excel_location
    def read_file(self):
        try:
            health_cost = pd.read_excel(self.excel_location,sheet_name = "By Family")
            print("File reads successfully")
        except:
            print("There's no sheet named By Family")
        return health_cost
    def clean_file(self):
        health_cost = self.read_file()
        health_cost_clean_col = perprocess.std_col_names(health_cost)
        return health_cost_clean_col
    def dict_file(self):
        health_cost_clean_col = self.clean_file()
        health_cost_dict = health_cost_clean_col.to_dict(orient="records")
        return health_cost_dict
```
